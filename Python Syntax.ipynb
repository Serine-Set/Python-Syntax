```python
# Welcome message
print("Welcome to Python Pizza Deliveries")

# Prompting user for their pizza size
size = input("What size pizza do you want? (S, M, L): ").upper()

# Prompting user for pepperoni choice
add_pepperoni = input("Do you want pepperoni? (Y/N): ").upper()

# Prompting user for extra cheese choice
extra_cheese = input("Do you want extra cheese? (Y/N): ").upper()

# Initial bill based on pizza size
bill = 0

if size == "S":
    bill += 15
elif size == "M":
    bill += 20
elif size == "L":
    bill += 25

# Adding pepperoni cost based on size
if add_pepperoni == "Y":
    if size == "S":
        bill += 2
    else:
        bill += 3

# Adding extra cheese cost
if extra_cheese == "Y":
    bill += 1

# Displaying the final bill
print(f"Your final bill is: ${bill}.")
```
